# Disney+ Catalog Insights

Analytics on the self-scraped Disney+ catalog in PostgreSQL.

**Run first:** `make up && make scrape --sample && make load && make views`

```python
import os
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine(
    "postgresql://{user}:{password}@{host}:{port}/{db}".format(
        user=os.getenv("POSTGRES_USER", "disney"),
        password=os.getenv("POSTGRES_PASSWORD", "disney"),
        host=os.getenv("POSTGRES_HOST", "localhost"),
        port=os.getenv("POSTGRES_PORT", "5432"),
        db=os.getenv("POSTGRES_DB", "disney_catalog"),
    )
)

summary = pd.read_sql("SELECT * FROM v_catalog_summary", engine)
summary
```

```python
franchises = pd.read_sql("SELECT * FROM v_franchise_stats ORDER BY titles DESC", engine)
franchises
```

## Headline Findings

1. **Franchise concentration** — Marvel, Disney, and Pixar dominate catalog depth
2. **Originals mix** — Disney+ Originals skew toward series (Mandalorian, Loki, Andor)
3. **Catalog growth** — Major launch window around Disney+ debut (Nov 2020) with steady originals drops after

```python
import plotly.express as px

growth = pd.read_sql("SELECT * FROM v_catalog_growth", engine)
px.line(growth, x="month_added", y="titles_added", markers=True, title="Catalog Growth Over Time")
```
